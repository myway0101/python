## 조건문 
조건문.png

1. `if` 문은 반드시 일정한 참/거짓을 판단할 수 있는 `조건식`과 함께 사용이 되어야한다.
`if <조건식>:`

2-1. `<조건식>`이 참인 경우 `:` 이후의 문장을 수행한다.

2-2. `<조건식>`이 거짓인 경우 `:` 이후의 문장을 수행하지 않는다.

* 이때 반드시 **들여쓰기를** 유의해야한다. 

 - 파이썬에서는 코드 블록을 자바나 C언어의 `{}`와 달리 `들여쓰기`로 판단하기 때문이다.
들여쓰기 방식을 통일해서 써야 한다.
 - 들여쓰기가 중요한 문법이다. 틀리면 오류가 발생한다.
 - 앞으로 배울 if, else, for, def 등의 문법들의 끝이 콜론(:)으로 끝나는데, 이경우 아래부분에 들여쓰기가 반드시 필요하다.

* ex) 
      if 조건식:
         print()
         print()
     
         
조건식.png       
         
 - 들여쓰기 유의사항
  - 동일한 블록 내에서 들여쓰기는 모두 동일해야 한다.
  - 들여쓰기를 tab 또는 space를 몇번할지는 규칙은 없으나 같은 블록내에서 동일하게 해야 한다.
  - 일반적으로 콜론(:) 다음 엔터를 입력하면 자동으로 (space 4칸) 들여쓰기가 된다.

```python
a = 9
print(a==9)
if a == 9:
  print("a는 9 입니다")
```

```python
age = 25
if age < 20:
  print("청소년 할인")
else:
  print("정상 가격 적용")

```

```python
number = [1, 5, 7]
print(1 in number)
if 3 in number:
  print("number에 3이 있습니다.")
else:
  print("number에 3이 없습니다.")
```

## 복합조건식
- 더 정교한 조건을 걸어주기 위해 조건 연산자와 논리 연산자를 조합


` if (  ) and (  ):`

` if (  ) or (  ):`


```python
a = 6
b = 10
if (a > 5) and (b > 5):
  print("a와 b 모두 5보다 큽니다.")
```

```python
a = 5
b = 10
if (a > 5) or (b > 5):
  print("적어도 a와 b 둘 중 하나는 5보다 큽니다.")
```

### `break`

break 문은, 가장 가까이서 둘러싸는 for 나 while 루프로부터 빠져나가게 만듭니다.

```python
for i in range(100):    # 0부터 99까지 반복
    print(i)
    if i == 10:    # i가 10일 때
        break       # 반복문을 끝냄. for의 제어흐름을 벗어남
```

### `continue` 

continue 문은 continue 이후의 코드를 수행하지 않고 다음 요소를 선택해 반복을 계속 수행합니다.

```python
for i in range(10):   
    if i == 5:    # i가 5일 때
        continue       # 아래코드를 수행하지 않고 반복 수행
    print(i)
```

```python
for i in range(10):   
    if i == 5:    # i가 5일 때
        pass      # pass 
    print(i)
```

```python
for i in range(10):   
    if i == 5:    # i가 5일 때
        pass      # pass 는 아무것도 실행하지 않고 패스
    else:
      print(i)
```
