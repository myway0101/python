##  `Range`(레인지)

레인지는 숫자의 시퀀스를 나타내기 위해 사용됩니다.

기본형 : `range(n)` 


> 0부터 n-1까지 값을 가짐


범위 지정 : `range(n, m)` 

> n부터 m-1까지 값을 가짐

범위 및 스텝 지정 : `range(n, m, s)`

> n부터 m-1까지 +s만큼 증가한다

````python
print(range(10))
print(type(range(10)))
````

````python
print(list(range(10)))
````

````python
print(list(range(1, 10)))
````

````python
print(list(range(1, 10, 3)))
````

````python
print(list(range(10, 0, -2)))
````

## 시퀀스에서 활용할 수 있는 연산자/함수 

|operation|설명|
|---------|---|
|x in s	|containment test|
|x not in s|containment test|
|s1 + s2|concatenation|
|s * n|n번만큼 반복하여 더하기
|s[i]|indexing|
|s[i:j]|slicing|
|s[i:j:k]|k간격으로 slicing|
|len(s)|길이|
|min(s)|최솟값|
|max(s)|최댓값|
|s.count(x)|x의 갯수|

````python
l = [1,2,3,4,5]
print(3 in l)
print(3 not in l)
print(len(l))
print(max(l))
print(min(l))
print(l[0])
print(sum(l))
````

# set(셋), dictionary(딕셔너리)

* `set`과 `dictionary`는 기본적으로 순서가 없습니다.

## `set`

세트는 수학에서의 집합과 동일하게 처리됩니다. 

세트는 중괄호`{}`를 통해 만들며, 순서가 없고 중복된 값이 없습니다.

**활용법**
```python
{value1, value2, value3}
```

|연산자/함수|설명|
|---|---|
|a - b|차집합|
|a \| b|합집합|
|a & b|교집합|
|a.union(b)|합집합|
|a.intersection(b)|교집합|

````python
a = {2, 3, 1}
b = {3, 6, 9}
print(a)
print(b)
````

````python
print(a[0])
````

````python
print(a-b)
print(a|b)
print(a&b)
print(a.union(b))
````

````python
a = [1, 2, 3, 2, 1, 2, 3]
print(list(set(a)))
````

````python
a = set("hello")
print(a)
````

````python
a = {1, 2, 3}
a.add(4)
a.remove(2)
print(a)
````

## `dictionary`

**활용법**
```python
{Key1:Value1, Key2:Value2, Key3:Value3, ...}
```
* 이러한 대응 관계를 나타낼 수 있는 자료형을 연관 배열(Associative array) 또는 해시(Hash)라고 한다.

* 딕셔너리(사전)는 `key`와 `value`가 쌍으로 이뤄져있으며, 매우 유용한 파이썬 자료구조입니다. 

* "people"이라는 단어에 "사람", "baseball"이라는 단어에 "야구"라는 뜻이 부합되듯이 딕셔너리는 Key와 Value를 한 쌍으로 갖는 자료형이다. 예컨대 Key가 "baseball"이라면 Value는 "야구"가 될 것이다.

* 딕셔너리는 리스트나 튜플처럼 순차적으로(sequential) 해당 요솟값을 구하지 않고 Key를 통해 Value를 얻는다. 이것이 바로 딕셔너리의 가장 큰 특징이다.

* baseball이라는 단어의 뜻을 찾기 위해 사전의 내용을 순차적으로 모두 검색하는 것이 아니라 baseball이라는 단어가 있는 곳만 펼쳐 보는 것이다.

* `{}`를 통해 만들며, `dict()`로 만들 수도 있습니다.

* 리스트나 튜플, 문자열은 요솟값을 얻고자 할 때 인덱싱이나 슬라이싱 기법 중 하나를 사용했다. 
* 하지만 딕셔너리는 단 한 가지 방법뿐이다. 바로 Key를 사용해서 Value를 구하는 방법이다

````python
phone_book = {"서울":"02", "경기":"031"}
print(phone_book)
print(type(phone_book))
````

````python
print(phone_book["서울"])
````

````python
phone_book["전남"] = "061"
print(phone_book)
````

````python
phone_book["서울"] = "01"
print(phone_book)
````

````python
print(phone_book.keys())
````

````python
print(phone_book.values())
````

````python
del phone_book["서울"]
print(phone_book)
````

# 파이썬 기본 자료구조 (리스트, 튜플, 셋, 딕셔너리) 분류

- 관리대상
 - 값만 관리: 리스트, 튜플, 셋(집합)
 - 키와 값의 쌍 관리: 딕셔너리

- 순서 관리 여부
 - 가능: 리스트, 튜플
 - 불가능: 딕셔너리, 집합
 
- 수정 가능 여부
 - 가능: 리스트, 셋, 딕셔너리
 - 불가능: 튜플
 
- 중복 가능 여부
 - 가능: 리스트, 튜플, 딕셔너리의 값
 - 불가능: 셋, 딕셔너리의 키
